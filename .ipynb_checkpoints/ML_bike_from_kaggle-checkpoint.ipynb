# Random Forest analysis of bike data from kaggle

In this notebook I am fitting a random forest model to the bike data from this popular kaggle data set
https://www.kaggle.com/c/bike-sharing-demand . The data is bike rental counts over time for both registered and casual users, a variety of weather data and whether it's a working day or a holiday.

First, I will look at the data and adjust the features as needed, explaining as I go. Next I will perform a random forest model on the unadjusted data and discuss the results. Then I'll also look at the data I adjusted based on the longitudinal analysis of the seasonality and trend.The following are the variables provided in the csv files:

**datetime:**   date and hour in "mm/dd/yyyy hh:mm" format

**season:**     Four categories-> 1 = spring, 2 = summer, 3 = fall, 4 = winter

**holiday:**    whether the day is a holiday or not (1/0)

**workingday:** whether the day is neither a weekend nor holiday (1/0)

**weather:**    Four Categories of weather

1. Clear, Few clouds, Partly cloudy, Partly cloudy

2. Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist

3. Light Snow and Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds

4. Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

**temp:**       hourly temperature in Celsius

**atemp:**      "feels like" temperature in Celsius

**humidity:**   relative humidity

**windspeed:**  wind speed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
%matplotlib inline
```

import the csv file and look at the variables. There are several variables that are int64 type, but should be categorical. These include season, holiday, workingday, and weather. Also we'll extract hour, month and day of week and coerce those to categorical data type as well.

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d %H:%M:%S')
data = pd.read_csv('Bike_train.csv', parse_dates=['datetime'], index_col='datetime',date_parser=dateparse)
print(data.info())
```

## Feature engineering: 


```python
# add columns for hour, month and day of week
data['Hour'] = data.index.hour
data['Month'] = data.index.month
data['day of week'] = data.index.dayofweek

#coerce into data type: category
categoryVars = ["Hour","day of week","Month","season","weather","holiday","workingday"]
for var in categoryVars:
    data[var] = data[var].astype("category")
    
data.head(1)
```

## windspeed and humidity - 

Lots of people in this challenge used random forest regression or knn to impute the windspeed zero values. I tried that with no success (predicting known windspeed values after train test split showed very poor preditions for windspeed from the other features.) Here I am choosing to leverage the fact that we have longitudinal data to assume continuity of windspeed over time. If we plot the windspeed as a function of time, we see that it is very noisy, but that most of the zeros are discontinuous, so perhaps back-filling will give a more accurate result.

It turns out that the humidity is zero for one day in March 2011, which appears to be an error, so I will replace those zeros with the humidity from the day before, but none of the test data have this problem, so it probably doesn't make much of a difference. I could also just drop those data without much trouble.

```python
plt.figure(figsize=(16,5))
data['windspeed']['2011-06'].plot()
windfill = data['windspeed'].replace(to_replace=0,method='bfill')
data['windfill']=windfill
data['windfill']['2011-06'].plot()
```

```python
plt.figure(figsize=(16,5))
data['humidity']['2011-03'].plot()
humidfill = data['humidity'].replace(to_replace=0,method='ffill')
data['humidfill']=humidfill
data['humidfill']['2011-03'].plot()
```

Now to consider what features to use in what models and also what models to use:

First, I will think about the casual users vs registered users. The holiday and workingday features are redundant unless we consider that the registered users seem to be commuting (see longitudinal analysis notebook with time of day) and the casual users seem to be on the weekends. So I'll use the holiday info for the casual users and the workingday for the registered.

After looking at the temp vs atemp info, I can't imagine why they are both included. temp correlates ever so slightly more with the casual and registered counts than atemp, so I'm going to use that and drop atemp. I also feel like season and month conflict, so I'll try them both separatelyand use whatever works better.
